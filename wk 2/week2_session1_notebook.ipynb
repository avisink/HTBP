### 🔴 CRITICAL CONCEPT: `input()` ALWAYS Returns a String!

No matter what the user types, `input()` gives you a **string** (text).

Even if they type a number, it's a string!

````python
# Let's prove it!
age = int(input("How old are you? "))   # Try typing "25"

print("You entered:", age)
print("The type is:", type(age))

birth_year = 2026 - age
print(birth_year)
````

## Part 3: Type Conversion - Changing Data Types

Python provides functions to convert between data types:

| Function | Converts To | Example |
|----------|-------------|----------|
| `int()` | Integer (whole number) | `int("25")` → `25` |
| `float()` | Float (decimal) | `float("3.14")` → `3.14` |
| `str()` | String (text) | `str(42)` → `"42"` |
| `bool()` | Boolean (True/False) | `bool(1)` → `True` |

### String to Integer: `int()`

````python
# Convert string to integer
age_str = "25"
print("Before conversion:", type(age_str))

age_int = int(age_str)
print("After conversion:", type(age_int))

# Now we can do math!
next_year = age_int + 1
print("Next year you'll be:", next_year)
````

### String to Float: `float()`

````python
# Convert string to float
price_str = "19.99"
print("Before:", type(price_str))

price_float = float(price_str)
print("After:", type(price_float))

# Calculate tax
tax = price_float * 0.08
total = price_float + tax
print(f"Price: ${price_float}")
print(f"Tax: ${tax:.2f}")
print(f"Total: ${total:.2f}")
````

### Number to String: `str()`

````python
# Convert number to string for concatenation
score = 95
total = 100

# This would cause an error:
# message = "You got " + score + " out of " + total  # ERROR!

# Solution: convert numbers to strings
message = "You got " + str(score) + " out of " + str(total) # like, you got so so score out of the total
print(message) # print the message

# Or use commas (easier!):
print("You got", score, "out of", total)
````

### Combining Input and Conversion

````python
# Get a number from user and convert it immediately
age = int(input("Enter your age: "))  # Convert to int right away!

# Now we can do math
months_old = age * 12
next_year_age = age + 1

print("\n--- Age Calculator ---")

# the \n in a print statement is used to print stuff on a new line
# to keep your code clean and legible

print("You are approximately", months_old, "months old")
print("Next year you will be", next_year_age)
````

````python
# Get a decimal number from user
height = float(input("Enter your height in meters: "))

# Convert to feet (1 meter = 3.28084 feet)
height_feet = height * 3.28084

print(f"Your height is {height_feet:.2f} feet")

# you can format variables inside your print statements  using curly braces
````

### ⚠️ Common Conversion Errors

````python
# These will cause errors - uncomment one at a time to see:

# 1. Can't convert text to a number
# age = int("hello")  # ValueError: invalid literal for int()

# 2. Can't convert a decimal string to int directly
# num = int("3.14")  # ValueError
# Solution: convert to float first, then to int
num = int(float("3.14"))  # This works! Result: 3
print(num)

# 3. Be careful with user input!
# If user types "twenty" instead of "20", conversion fails
# We'll learn how to handle this in Week 5!
````

Now let's combine everything to make useful programs!

### Example 1: Interactive Greeter

````python
# Get user information
name = input("What's your name? ")
age = int(input("How old are you? "))
city = input("Where are you from? ")

# Process the information
birth_year = 2026 - age
name_length = len(name)

# Display results
print("\n--- User Profile ---")
print("Name:", name.upper())
print("Age:", age, "years old")
print("Born approximately:", birth_year)
print("Hometown:", city)
print("Your name has", name_length, "letters")
print("-------------------")
````

### Example 2: Simple Calculator

````python
# Get two numbers from the user
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Perform calculations
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2

# Display results
print("\n=== Calculator Results ===")
print(f"{num1} + {num2} = {addition}")
print(f"{num1} - {num2} = {subtraction}")
print(f"{num1} × {num2} = {multiplication}")
print(f"{num1} ÷ {num2} = {division:.2f}")
print("=========================")
````

### 💡 Quick Tip: F-Strings (Formatted String Literals)

F-strings are a modern, clean way to format strings in Python. Put an `f` before the quote and use `{}` to insert variables:

```python
name = "Alice"
age = 20

# Old way (concatenation):
print("My name is " + name + " and I am " + str(age) + " years old")

# New way (f-string):
print(f"My name is {name} and I am {age} years old")
```

F-strings automatically convert numbers to strings and are much easier to read!

````python
# F-string examples
name = "Bob"
age = 25
height = 1.75
score = 95.5

# Basic f-string
print(f"Hello, {name}!")

# Multiple variables
print(f"{name} is {age} years old")

# Expressions inside {}
print(f"Next year {name} will be {age + 1}")

# Format decimals with :.2f (2 decimal places)
print(f"Height: {height:.2f} meters")
print(f"Score: {score:.1f}%")
````
